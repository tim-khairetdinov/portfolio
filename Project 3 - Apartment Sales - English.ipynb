# Studying Apartment Sales Postings

Yandex.Apartments has provided us the data on apartment sales postings in St. Petersburg and the greater region over the past several years. They are looking for a way to determine the market price for a given apartment. Our task is find the parameters that affect what an acceptable market price is. This will help the client create an automated system for tracking anomalies and fraud.

There are two types of data available to us. The first is information entered by the user, such as the listed price. The second is generated automatically by the system, such as distance to the nearest park.

Note: in the Russian Federation, an **"apartament"** is an apartment created out of a formerly non-residential space, usually a converted industrial space. These are covered by a different set of regulations than apartments in general.

## Data Preprocessing

### Column titles

### Missing Values

Let's take a look at our missing values.

#### Miscellaneous Missing Values

`floors_total` (number of floors in the building) and `locality_name` (name of the town the listing is in) have a very low number of missing values (<0.5%). We can safely discard these.

```python
data.dropna(subset=['floors_total', 'locality_name'], inplace=True)

data.info()
```

#### ceiling_height

`ceiling_height` is the height of the listing's ceiling in meters. Let's see what options we have here.

```python
data['ceiling_height'].sort_values().unique()
```

We can see several problems right away:

1. We have missing values (na)
2. We have anomalously low values
3. We have anomalously high values.

Let's start with the low values. According to the Building Code of the Russian Federation, the minimum ceiling height in living spaces is 2.5 m (https://potolki-info.ru/kvartira/standartnaya-vysota-potolkov-normy-i-praktika). For "apartaments" (converted industrial spaces, a different real estate category in the RF), there isn't an official minimum. Therefore, we're going to select listings with `ceiling_height` of under 2.5, and drop them if `is_apartment` is false. 

```python
data.query('ceiling_height < 2.5 and is_apartment != True')
```

73 entries. Let's drop them.

```python
data = data[~((data['ceiling_height']<2.5)&(data['is_apartment']!=True))]

data['ceiling_height'].sort_values().unique()
```

Excellent, we got rid of the anomalously low values. Let's now take a look at the anomalously high values. What's the dividing line between "high" and "anomalously high"?

"The maximum ceiling height is not generally regulated, but it's usually not greater than 3.2 m in modern buildings. Older buildings with larger dimensions can reach 4 m or above" (https://potolok-exp.ru/montazh-potolkov/standartnaya-vysota-potolkov.html)

4 m is pretty high, but let's set it as acceptable. Let's cut off everything above 5 m.

Spme of the values above 5 m have clear problems with a misplaced decimal point. Everything about 10 is probably actually 1 m. But we need to keep in mind that everything below 2.5 m is illegal anyway. Thus, we need to get rid of all entries with `ceiling_height` between 5 and 25, and divide everything above 25 by 10.

```python
data[(data['ceiling_height'] > 5) & (data['ceiling_height'] < 25)]['ceiling_height'].count()
```

There are only 16 of these, so we can safely discard them.

```python
#drop the illegal values
data = data[~((data['ceiling_height'] > 5) & (data['ceiling_height'] < 25))]

#drop the values of 100, it's too large
data = data[~(data['ceiling_height'] == 100)]

#check results
data['ceiling_height'].sort_values().unique()
```

```python
#divide the high values by 10
#make a temporary ceiling_fixed column, copy ceiling_height there if it's <5, and divide it by 10 and copy if not
data['ceiling_fixed'] = data['ceiling_height'].apply(lambda x: x / 10 if x > 5 else x)

#check results
data[data['ceiling_height'] > 4].head(5)
```

The new column works well, so we can replace `ceiling_height` with it.

```python
#delete the old column
data.drop('ceiling_height', axis=1, inplace=True)

#rename the new column
data = data.rename(columns={'ceiling_fixed': 'ceiling_height'})

#check results
data['ceiling_height'].sort_values().unique()
```

Excellent, we fixed the anomalously low and anomalously high values. What about the missing ones?

We can't assume that these missing values are data entry errors. If the user did not input the ceiling height when they made the posting, it would remain empty. This means that these are legitimate entries. However, NaNs will cause problems when we're building graphs at doing other analysis. In order to fix this, we will fill the NaNs with -2, so we wouldn't confuse them with actual entries.

```python
#fill NaNs
data['ceiling_height'] = data['ceiling_height'].fillna(-2)

#test results
data.info()
```

Missing values in `ceiling_height` are done.

"The accounting rate is the mininum amount of living space per person. In St. Petersburg, the following accounting rates are currently active: 9 sq.m of total living space for those living in dedicated apartments and detached houses; 15 sq. m of total living space for those living in communal apartments" (https://gu.spb.ru/help/usefulinfo/Socialnyj-naem/) 

This means that our jump at 18 sq. m is single-bedroom apartments for two people.

Let's see if we have any listings with less than 9 sq. m of living space, which would be illegal.

That's not good. Let's delete these illegal values.

```python
#drop values
data = data[~(data['living_area'] < 9)]

#test results
data[data['living_area'] < 9]['living_area'].count()
```

```python
data['living_area'].sort_values(ascending=False).head(15)
```

However, just like with `ceiling_height`, we can't leave them as NaN, since that would mess up our graphs, so we will set them up as zeroes.

### Removing Duplicates

Alright, our `locality_name` is a string. This could lead to problems with upper case. Let's look at the column in detail.

That's far too much data to fix by hand, so we're going to have to lemmatize. Our assumption for this is that only the actual name of the population center is important, while the type (village, town, etc.) does not matter. This means we can drop types from the lemmatized list.

```python
def lemmatizer (text):
    
    #get the lemmas from the text into a tuple
    lemmas = m.lemmatize(text)
    
    #convert the tuple into a list
    lemmas = list(lemmas)
        
    #build the list of all the unnecessary elements (population center types, spaces, \n)
    extras = ['городской', 'поселок', 'тип', 'деревня', 'садовой', 'товарищество', 'некоммерческий',
             'при', 'станция', 'железнодорожный', 'село', 'коттеджный', ' ', '\n']
    
    #drop the unnecessary elements from the lemmas list
    for entry in extras:
        lemmas = list(filter((entry).__ne__, lemmas))
    
    #bring the list back to a string and return it
    return (' '.join(lemmas))
```

```python
%%time
data['lemmatized_location'] = data['locality_name'].apply(lemmatizer)

data.head(10)
```

Still not done. Let's create two temporary columns:

- `kitchen_subtract` = `total_area` - `kitchen_area`
- `living_subtract` = `total_area` - `living_area`

Our `kitchen_subtract` shouldn't be under 9.96, while `living_subtract` shouldn't be under 6.96 (don't forget about the bathroom). All the lines with values below those can be safely discarded as illegal.

```python
data['kitchen_subtract'] = data['total_area'] - data['kitchen_area']
data['living_subtract'] = data['total_area'] - data['living_area']

#put it together into a temporary table
data[data['total_area'] < 25][['total_area', 'kitchen_area', 'kitchen_subtract',
                               'living_area', 'living_subtract',
                               'is_apartment', 'studio']].sort_values(by=['kitchen_subtract', 'living_subtract'])
```

Wait, take a look at line with id 1286. 18.2 sq. m kitchen_area, 18 sq. m living_area, coming together to make... 24.6 sq. m? Those are some magical meters.

Alright, let's make a `total_area_red_flag` column, which is False by default, but is set to True if:

- `kitchen_subtract` < 9.96 (not enough space for living area and bathroom), or
- `living_subtract` < 6.96 (not enough space for the kitchen and bathroom), or
- `kitchen_area` + `living_area` + 0.96 >= `total_area` (lying about the kitchen or living area).

```python
#the flag setting function
def total_red_flagger (line):
    
    #temporary variables
    k_sub = line['kitchen_subtract']
    l_sub = line['living_subtract']
    t_off = line['total_area']
    t_act = line['kitchen_area'] + line['living_area'] + 0.96
    
    #give values
    #not enough space for living space and a bathroom
    if (k_sub < 9.96):
        return True
    
    #not enough space for a kitchen and a bathroom
    elif (l_sub < 6.96):
        return True
    
    #lying about the total area
    elif (t_act > t_off):
        return True
    
    #if everything else is fine, set it to False
    else:
        return False
```

```python
#apply the function
data['total_area_red_flag'] = data.apply(total_red_flagger, axis=1)

data[data['total_area'] < 25][['total_area', 'kitchen_area', 'living_area',
                               'is_apartment', 'total_area_red_flag']].sort_values(by='total_area_red_flag',
                                                                                            ascending=False)
```

There's our excuse to drop more line. Remember, SNiP doesn't apply to apartaments, so if the line has `is_apartment` set to True, then we can't drop it.

```python
data = data[~((data['total_area_red_flag'] == True) & (data['is_apartment'] == False))]

data[data['total_area'] < 25][['total_area', 'kitchen_area', 'living_area',
                               'is_apartment']].sort_values(by='total_area')
```
